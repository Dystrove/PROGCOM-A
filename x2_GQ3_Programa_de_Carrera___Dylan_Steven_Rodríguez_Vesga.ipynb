```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# FunciÃ³n de detecciÃ³n de anemia
def detectar_anemia(edad, genero, hemoglobina):
    if edad >= 12:
        if genero == "H":
            return hemoglobina < 13
        else:
            return hemoglobina < 12
    else:
        return hemoglobina < 11

# Widgets de entrada
edad_input = widgets.IntSlider(value=10, min=0, max=100, description='Edad:', style={'description_width': 'initial'})
genero_input = widgets.ToggleButtons(options=[('Hombre', 'H'), ('Mujer', 'M')], description='GÃ©nero:')
hemo_input = widgets.FloatSlider(value=11.0, min=5, max=20, step=0.1, description='Hemoglobina (g/dl):', style={'description_width': 'initial'})

# BotÃ³n y salida
boton = widgets.Button(description="Evaluar anemia", button_style='info')
salida = widgets.Output()

# FunciÃ³n para mostrar resultado
def evaluar_anemia(b):
    with salida:
        clear_output()
        edad = edad_input.value
        genero = genero_input.value
        hemoglobina = hemo_input.value

        tiene_anemia = detectar_anemia(edad, genero, hemoglobina)

        if tiene_anemia:
            print("ðŸ”´ Resultado: Tiene anemia")
            print("RecomendaciÃ³n: Consultar con un profesional de la salud.")
        else:
            print("ðŸŸ¢ Resultado: No tiene anemia")
            print("El nivel de hemoglobina estÃ¡ dentro del rango normal.")

# Asignar funciÃ³n al botÃ³n
boton.on_click(evaluar_anemia)

# Mostrar interfaz
display(widgets.VBox([
    widgets.HTML("<h3>ðŸ©¸ Evaluador de Anemia</h3>"),
    edad_input,
    genero_input,
    hemo_input,
    boton,
    salida
]))
```
