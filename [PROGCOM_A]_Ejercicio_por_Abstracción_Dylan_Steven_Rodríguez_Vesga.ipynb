```python
from abc import ABC, abstractmethod

class PersonaCredito(ABC):
    def _init_(self, edad, tiene_tarjeta, sin_morosidad, contrato_fijo,
                 contrato_estable, tuvo_credito_otro_banco, tuvo_credito_aca):
        self.edad = edad
        self.tiene_tarjeta = tiene_tarjeta
        self.sin_morosidad = sin_morosidad
        self.contrato_fijo = contrato_fijo
        self.contrato_estable = contrato_estable
        self.tuvo_credito_otro_banco = tuvo_credito_otro_banco
        self.tuvo_credito_aca = tuvo_credito_aca

    @abstractmethod
    def calcular_puntos(self):
        pass

    @abstractmethod
    def aprobar_credito(self):
        pass


class SistemaEvaluacion(PersonaCredito):
    def calcular_puntos(self):
        puntos = 0
        if self.tiene_tarjeta:
            puntos += 2
        if self.sin_morosidad:
            puntos += 3
        if self.contrato_fijo:
            puntos += 2
        if self.contrato_estable:
            puntos += 3
        if self.tuvo_credito_otro_banco:
            puntos += 2
        if self.tuvo_credito_aca:
            puntos += 3

        if 18 <= self.edad <= 24:
            puntos += 2
        elif 25 <= self.edad <= 35:
            puntos += 4
        elif 36 <= self.edad <= 50:
            puntos += 3
        else:
            puntos += 1

        return puntos

    def aprobar_credito(self):
        return self.calcular_puntos() >= 7


def main():
    cantidad = int(input("¿Cuántas personas deseas evaluar?: "))
    aprobados = 0

    for i in range(cantidad):
        print(f"\n🧾 Evaluación de la persona #{i+1}")
        edad = int(input("Edad: "))
        tiene_tarjeta = input("¿Tiene tarjeta de otro banco? (s/n): ").lower() == 's'
        sin_morosidad = input("¿No está reportado negativamente? (s/n): ").lower() == 's'
        contrato_fijo = input("¿Tiene contrato fijo? (s/n): ").lower() == 's'
        contrato_estable = input("¿Tiene contrato indefinido? (s/n): ").lower() == 's'
        tuvo_credito_otro_banco = input("¿Tuvo crédito en otro banco? (s/n): ").lower() == 's'
        tuvo_credito_aca = input("¿Tuvo crédito en esta entidad? (s/n): ").lower() == 's'

        persona = SistemaEvaluacion(edad, tiene_tarjeta, sin_morosidad,
                                    contrato_fijo, contrato_estable,
                                    tuvo_credito_otro_banco, tuvo_credito_aca)

        puntos = persona.calcular_puntos()
        if persona.aprobar_credito():
            print(f"✅ Crédito aprobado (Puntaje: {puntos})")
            aprobados += 1
        else:
            print(f"❌ Crédito no aprobado (Puntaje: {puntos})")

    print(f"\n📊 Total de personas aprobadas: {aprobados} de {cantidad}")

if _name_ == "_main_":
    main()
```
